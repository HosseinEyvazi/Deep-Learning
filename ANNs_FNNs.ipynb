## Introduction to Artificial Neural Networks (ANNs)

Artificial Neural Networks (ANNs) are computational models inspired by the human brain's structure and function. They consist of interconnected nodes, or **neurons**, that process data by responding to external inputs and relaying information between each other. ANNs are designed to recognize patterns, make decisions, and solve complex problems by learning from data.

**Key Points:**
- **Biological Inspiration:** Mimic the behavior of biological neural networks.
- **Data Processing:** Handle inputs like x₁, x₂, x₃, multiply them by weights, and pass them through layers.
- **Capacity and Overfitting:** More layers increase the number of weights, enhancing the model's capacity to learn but also increasing the risk of overfitting.
- **Structure:** Comprises input, hidden, and output layers.
- **Layers vs. Nodes:** The number of layers refers to the number of edges (connections), not the nodes (neurons) within them.

**Industry Insight:**
Approximately **70%** of business projects involve machine learning, while the remaining **30%** focus on deep learning. Deep learning models often require substantial computational resources, such as GPUs, and consume more memory compared to traditional machine learning models, which are typically faster and more memory-efficient.

## Importance of Non-Linearity

Non-linearity is crucial in ANNs as it allows the network to learn and model complex relationships within the data. Without non-linear activation functions, the network would behave like a simple linear model, limiting its ability to solve intricate problems.

## Activation Functions

Activation functions introduce non-linearity into the network, enabling it to learn complex patterns. They determine whether a neuron should be activated based on the input it receives.

### 1. ReLU (Rectified Linear Unit)

- **Definition:** ReLU(x) = max(0, x)
- **Advantages:**
  - Most powerful in practice for hidden layers.
  - Simple and efficient to compute.
  - Helps mitigate the vanishing gradient problem.
- **Derivative:** ReLU is differentiable for all x except at x = 0, which simplifies gradient calculations.

### 2. Sigmoid

- **Definition:** Sigmoid(x) = 1 / (1 + e^(-x))
- **Advantages:**
  - Outputs values between 0 and 1, suitable for probability estimation.
- **Issues:**
  - **Vanishing Gradients:** Derivatives approach zero for extreme input values, slowing down learning.
  - Not zero-centered, which can affect convergence.

### 3. Tanh (Hyperbolic Tangent)

- **Definition:** Tanh(x) = 2 / (1 + e^(-2x)) - 1
- **Advantages:**
  - Outputs values between -1 and 1, making it zero-centered.
- **Issues:**
  - Similar to sigmoid, suffers from vanishing gradients.

### 4. Softmax

- **Definition:** Converts a set of values into probabilities that sum to 1.
- **Usage:** Typically used in the output layer for multi-class classification problems.

### 5. Other Activation Functions

- **Leaky ReLU:** Allows a small, non-zero gradient when the unit is not active.
- **ELU (Exponential Linear Unit):** Smooths out ReLU by allowing negative outputs.
- **Swish:** A self-gated activation function defined as Swish(x) = x * Sigmoid(x).

### Choosing Activation Functions

- **Hidden Layers:** ReLU is generally preferred due to its efficiency and performance.
- **Output Layers:**
  - **Regression:** Linear activation (no activation function).
  - **Binary Classification:** Sigmoid activation.
  - **Multi-class Classification:** Softmax activation.

## Training Artificial Neural Networks

Training ANNs involves optimizing the network's parameters (weights and biases) to minimize a loss function, enabling the model to make accurate predictions.

### Gradient Descent

**Gradient Descent** is a fundamental optimization algorithm used to minimize the loss function by iteratively adjusting the model's parameters in the direction opposite to the gradient.

**Steps Involved:**

1. **Initialize Parameters:**
   - Randomly assign initial weights and biases.

2. **Forward Pass:**
   - Compute predictions by passing inputs through the network.

3. **Compute Loss:**
   - Evaluate the loss function (e.g., Mean Squared Error, Cross-Entropy).

4. **Backward Pass (Backpropagation):**
   - Calculate gradients of the loss with respect to each parameter.
   - Use the chain rule to propagate gradients backward through the network.

5. **Update Parameters:**
   - Adjust weights and biases using the gradients and a chosen learning rate (α):
     ```
     New Weight = Old Weight - (α * Gradient)
     ```

6. **Iterate:**
   - Repeat the process until convergence or a stopping criterion is met.

**Variants of Gradient Descent:**
- **Batch Gradient Descent:** Uses the entire dataset to compute gradients.
- **Stochastic Gradient Descent (SGD):** Uses one randomly selected sample for each update.
- **Mini-Batch Gradient Descent:** Uses a subset of the dataset for each update.

### Overfitting, Regularization, and Generalization
**Overfitting:**  
Occurs when a model learns the training data too well, capturing noise and losing the ability to generalize to new, unseen data.

**Regularization:**  
Techniques used to prevent overfitting by adding constraints or penalties to the model's complexity.

**Common Regularization Methods:**
- **L1 Regularization:** Adds the absolute value of weights as a penalty.
- **L2 Regularization:** Adds the squared value of weights as a penalty.
- **Dropout:**  
  Randomly deactivates neurons during training to prevent co-dependency.  
  **Explanation:**  
  Dropout is a regularization technique where, during each training iteration, a random subset of neurons is "dropped out," meaning they are temporarily removed from the network. This prevents the network from becoming too reliant on any specific neurons, encouraging the development of redundant representations and promoting robustness. By doing so, dropout helps reduce overfitting, as the model learns to generalize better to unseen data.
- **Early Stopping:** Stops training when performance on a validation set starts to degrade.

**Generalization:**  

The model's ability to perform well on new, unseen data (this means ROBUSTNESS) by capturing the underlying patterns rather than memorizing the training data.

### What is CUDA?

CUDA (Compute Unified Device Architecture) is a parallel computing platform and API developed by NVIDIA. It allows developers to leverage the power of NVIDIA GPUs for general-purpose processing. In the context of neural networks, CUDA accelerates matrix operations, which are fundamental to training and inference, providing significant speedups compared to CPU computations. While libraries like NumPy are excellent for numerical operations, they do not natively support GPU acceleration. CUDA bridges this gap, enabling efficient processing of large-scale neural network tasks.

### What is TPU?

TPU (Tensor Processing Unit) is a type of accelerator developed by Google specifically designed for machine learning workloads. TPUs are optimized for high-throughput matrix operations, making them ideal for training and deploying large neural networks. They offer faster computation times and higher efficiency compared to traditional GPUs, especially for tasks involving large-scale data and complex models.

In neural networks, where operations are heavily based on matrix multiplications, TPUs provide a significant performance boost. Libraries like TensorFlow are optimized to take full advantage of TPU capabilities, ensuring that models can be trained and deployed efficiently.

Pretrained models are widely used in deep learning for several reasons:

1. **High Number of Parameters**: Training models with millions (or billions) of parameters from scratch is computationally expensive.

## Transfer Learning
Transfer learning is a technique where a pretrained model is fine-tuned for a new task. This involves:

1. **Unfreezing Last Layers**: Instead of training the entire model, only the last few layers are unfrozen and trained on the new dataset. The earlier layers, which capture general features (like edges and textures), remain frozen.
2. **Example**: Use a pretrained AlexNet model for a new image classification task. Unfreeze the last few layers to adapt the model to the new dataset while keeping the initial layers frozen.
3. **Why It’s a Good Approach**: Transfer learning saves time and computational resources, especially when the new dataset is small. It also improves performance by leveraging knowledge learned from a large dataset.

# Keras Model Definitions: Sequential vs. Functional

Keras provides two main approaches to define neural networks: **Sequential** and **Functional** APIs. Each serves different purposes based on the complexity of your architecture.

## Sequential API
- **Simple, Linear Models**: Layers are stacked one by one from input to output.
- **When to Use**: Best for straightforward workflows with single inputs/outputs, no branching, and no intermediate layer access.
- **Advantages**: Easy to read and quick to implement for beginners.

## Functional API
- **Flexible Architectures**: Allows multiple inputs/outputs, layer reuse, and branching.
- **Advantages**:  
  - Access to intermediate layers for complex architectures.  
  - Ability to combine classification and regression outputs in a single model.  
  - Easier to implement transfer learning by branching and fine-tuning layers.
- **When to Use**: Ideal for multi-output models, multiple input types, or when fine-grained control over the flow of data is required.

---

## Pytorch setup for Regression (Price prediction)

````python
import torch
from torch.utils.data import DataLoader, Dataset, TensorDataset
import torch.nn as nn

# from torch.utils.data import

````

````python
# make handlable for CUDA (?)
X_train=torch.tensor(X_train)
X_test=torch.tensor(X_test)
y_train=torch.tensor(y_train)
y_test=torch.tensor(y_test)

#this is creating data loader for batching to solve ram limitation issue
train_loader=DataLoader(TensorDataset(X_train,y_train),batch_size=32,shuffle=True) # higher batch size makes model performance and speed higher. but problem is hardware llimitaion
test_loader=DataLoader(TensorDataset(X_test,y_test),batch_size=32,shuffle=True)
````

````python
class RegressionNN(nn.Module):
  def __init__(self,input_size):
    super().__init__()
    self.fc1=nn.Linear(input_size,64) # note that input size is equal to number of features not number of datapoints
    self.fc2=nn.Linear(64,32)
    self.fc3=nn.Linear(32,1) # task is regression that is why output layer is one

  def forward(self,x): # the 'forward' method is indeed special in PyTorch and has specific significance within the framework. When you define a neural network by subclassing nn.Module, the forward method is the method that defines how the input data flows through the network.
    x=torch.relu(self.fc1(x))
    x=torch.relu(self.fc2(x))
    x=self.fc3(x)
    return x
````

The `__setattr__` method in PyTorch's `nn.Module` plays a crucial role in registering layers when they are assigned. For example, when you define a layer like:

```python
self.fc1 = nn.Linear(input_size, 64)
```

the `__setattr__` method detects that `fc1` is an instance of `nn.Module` (in this case, an `nn.Linear` layer). It then calls:

```python
self.add_module("fc1", value)  # Registers nn.Linear as a submodule
```

This ensures that the layer is properly registered in `self._modules`, a dictionary that stores all submodules of the model. This mechanism allows PyTorch to manage layers, parameters, and buffers efficiently, enabling functionalities like model saving, loading, and device movement.

````python
def train_regression(train_loader, test_loader, model, criterion, optimizer, num_epochs):
    model.train()  # Set model to training mode at the start of each epoch
    train_losses , test_losses =[],[]
    for epoch in range(num_epochs):
        running_loss = 0.0
        # Training loop
        for inputs, targets in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs.float())
            loss = criterion(outputs.float(), targets.float())
            # train_losses.append(loss.item())
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
            # .item() is a PyTorch method that returns the value of the tensor as a Python number (float or int).
            #... Since loss is a scalar TENSOR, loss.item() retrieves that scalar VALUE as a normal Python float.


        # Average loss for the training data for this epoch
        train_loss = running_loss / len(train_loader)
        train_losses.append(train_loss)
        # Now evaluate on the test set
        model.eval()  # Set model to evaluation mode
        with torch.no_grad():  # Turn off gradient calculation (it's not needed during evaluation)
            test_loss = 0.0
            for inputs, targets in test_loader:
                outputs = model(inputs.float())
                loss = criterion(outputs.float(), targets.float())
                test_loss += loss.item()

            # Average loss for the test data for this epoch
            test_loss = test_loss / len(test_loader)
            test_losses.append(test_loss)
        # Print loss for both training and test data
        print(f'Epoch {epoch + 1}: Train loss = {train_loss:.4f}, Test loss = {test_loss:.4f}')
    plt.plot(train_losses, label='train loss')
    plt.plot(test_losses, label='test loss')
    plt.legend()
    plt.xlabel('epochs')
    plt.ylabel('loss')
    plt.grid()


pytorch_model=RegressionNN(X_train.shape[1])
criterion=nn.MSELoss()
optimizer=torch.optim.Adam(pytorch_model.parameters(), lr=.001)
# Call to train_regression assuming train_loader, pytorch_model, criterion, and optimizer are defined
# Call to train_regression assuming train_loader, test_loader, pytorch_model, criterion, and optimizer are defined
train_regression(train_loader, test_loader, pytorch_model, criterion, optimizer, num_epochs=10)
````

### Why Use `batch_size` in PyTorch?

- **Batch size** controls how many samples are processed at once in training.
- **Higher batch size** → More stable gradients, but **requires more memory** and may generalize poorly.
- **Lower batch size** → Noisier updates, but can escape sharp local minima and generalize better.

### Why Low `batch_size` in SGD May Cause Instability?

- Small batches add **noise** to gradients, helping escape poor minima.
- Too small can cause **flaring** (unstable convergence), while too large may get stuck in sharp minima.
- A **balance** is needed for stability and generalization. 🚀

### **Epochs vs. Batch Size**  

- **Epochs**: The number of times the entire dataset is passed through the model.  
- **Batch Size**: The number of samples processed before updating the model.  

To see all data once:  

Total Steps = Dataset Size/Batch Size
  
Larger batch sizes lead to fewer updates per epoch, while smaller batches provide more frequent updates. 🚀
